```python
import ast
import seaborn as sns
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gamma_list = ["0.1", "0.5", "0.9", "0.99", "0.995", "1"]
```

```python
# reward
plt.figure(figsize=(16, 8))
for gamma in gamma_list:
    q_logs = pd.read_csv('./data/'+gamma+'rewards_qlearning.csv', index_col=False).iloc[:, 1]
    avg_cumsum_rewards = q_logs.cumsum() / (np.arange(len(q_logs)) + 1)
    plt.plot(avg_cumsum_rewards, label=f"γ={gamma}")
plt.title("Q-Learning_reward")
plt.xlabel("Episode")
plt.ylabel("reward")
plt.legend()
plt.grid(True)
plt.show()

# TD error 
plt.figure(figsize=(16, 8))
for gamma in gamma_list:
    q_td_logs = pd.read_csv(f'./data/{gamma}td_error_qlearning.csv', index_col=False).iloc[:, 1]
    avg_td_error = q_td_logs.cumsum() / (np.arange(len(q_td_logs)) + 1)
    plt.plot(avg_td_error, label=f"γ={gamma}")
plt.title("Q-Learning_TD Error")
plt.xlabel("Episode")
plt.ylabel("Average TD Error")
plt.legend()
plt.grid(True)
plt.show()
```

```python
all_max_q_histories = []
all_mean_q_histories = []

for gamma in gamma_list:
    # CSV 불러오기
    max_q = pd.read_csv(f'./data/state_{gamma}maxQ_qlearning.csv', index_col=0).iloc[:, 0]
    mean_q = pd.read_csv(f'./data/state_{gamma}meanQ_qlearning.csv', index_col=0).iloc[:, 0]

    # 문자열 -> 리스트로 변환
    max_q = max_q.apply(ast.literal_eval)
    mean_q = mean_q.apply(ast.literal_eval)

    # 리스트 형태로 저장
    all_max_q_histories.append(list(max_q))
    all_mean_q_histories.append(list(mean_q))



# 9state 무작위 선택 
sampled_indices = random.sample(range(60), 9)

# Q-Max 그래
fig, axs = plt.subplots(3, 3, figsize=(18, 12))
axs = axs.flatten()

for i, state_idx in enumerate(sampled_indices):
    for gamma_idx, gamma in enumerate(gamma_list):
        max_q = all_max_q_histories[gamma_idx][state_idx]
        x = np.arange(0, len(max_q) * 100, 100)
        axs[i].plot(x, max_q, label=f"γ={gamma}")

    axs[i].set_title(f"State {state_idx} - Q Max")
    axs[i].set_xlabel("Episode")
    axs[i].set_ylabel("Q Max")
    axs[i].legend()
    axs[i].grid(True)

plt.tight_layout()
plt.show()

# Q-Mean
fig, axs = plt.subplots(3, 3, figsize=(18, 12))
axs = axs.flatten()

for i, state_idx in enumerate(sampled_indices):
    for gamma_idx, gamma in enumerate(gamma_list):
        mean_q = all_mean_q_histories[gamma_idx][state_idx]
        x = np.arange(0, len(mean_q) * 100, 100)
        axs[i].plot(x, mean_q, label=f"γ={gamma}")

    axs[i].set_title(f"State {state_idx} - Q Mean")
    axs[i].set_xlabel("Episode")
    axs[i].set_ylabel("Q Mean")
    axs[i].legend()
    axs[i].grid(True)

plt.tight_layout()
plt.show()

```

```python

# Q-Max - Q-Mean (Diff) 그래프 
fig, axs = plt.subplots(3, 3, figsize=(18, 12))
axs = axs.flatten()

for i, state_idx in enumerate(sampled_indices):
    for gamma_idx, gamma in enumerate(gamma_list):
        max_q = all_max_q_histories[gamma_idx][state_idx]
        mean_q = all_mean_q_histories[gamma_idx][state_idx]
        diff_q = np.array(max_q) - np.array(mean_q)
        x = np.arange(0, len(diff_q) * 100, 100)

        axs[i].plot(x, diff_q, label=f"γ={gamma}")

    axs[i].set_title(f"State {state_idx} - Q Max - Q Mean")
    axs[i].set_xlabel("Episode")
    axs[i].set_ylabel("Q Diff")
    axs[i].legend()
    axs[i].grid(True)

plt.tight_layout()
plt.show()

```

```python

all_diff_q_histories = []

for gamma in gamma_list:
    # CSV 불러오기
    diff_q = pd.read_csv(f'./data/state_{gamma}diffQ_qlearning.csv', index_col=0).iloc[:, 0]
    # 문자열 -> 리스트로 변환

    # 리스트 형태로 저장
    all_diff_q_histories.append(list(diff_q))


plt.figure(figsize=(12, 6))

for i, gamma in enumerate(gamma_list):
    diff_q_series = all_diff_q_histories[i]
    x = np.arange(0, len(diff_q_series) * 100, 100)  

    plt.plot(x, diff_q_series, label=f"γ={gamma}")

plt.title("Total Q-Max - Q-Mean")
plt.xlabel("Episode")
plt.ylabel("Sum of Q Diff")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
